```python
import pandas as pd

import matplotlib.pyplot as plt 
import folium as flm


data = pd.read_csv(r'C:\Users\PC\Downloads\dataset.csv')
carte = flm.Map(location =[data["Latitude"].mean(),data["Longitude"].mean()])
    
latitude = data["Latitude"].tolist()
longitude = data["Longitude"].tolist()

limit = 1000 
count = 0

for i in range(len(latitude)):
    if count < limit:
        marker=flm.Marker(location = [latitude[i], longitude[i]])
        marker.add_to(carte)
        count +=1
carte

    
```

```python
dataVie = data["FacilityType"].value_counts()

plt.figure(figsize=(10,10))
plt.title("tableau de valeurs")
plt.pie(dataVie,labels=dataVie.index)
```
